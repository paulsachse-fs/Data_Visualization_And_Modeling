## Dataset 1: Movie Reviews

##### Load Data

```python
import nltk
from nltk.corpus import movie_reviews

for pkg in ["movie_reviews", "punkt"]:
    try:
        nltk.data.find(f"corpora/{pkg}")
    except LookupError:
        nltk.download(pkg)

print("movie_reviews files:", len(movie_reviews.fileids()))
```

##### Create Dataframe and Clean

```python
import pandas as pd
import string
import re

def simple_clean(text):
    if not isinstance(text, str):
        return ""
    t = text.lower()
    t = t.translate(str.maketrans({c: " " for c in string.punctuation}))
    t = t.translate(str.maketrans({c: " " for c in string.digits}))
    t = re.sub(r"\s+", " ", t).strip()
    return t

rows = []
for fid in movie_reviews.fileids():
    cats = movie_reviews.categories(fid)
    if not cats:
        continue
    label = cats[0]       # pos or neg
    text = movie_reviews.raw(fid)
    rows.append({"text": text, "label": label, "dataset": "movie_reviews"})

mv = pd.DataFrame(rows)
mv["text_clean"] = mv["text"].map(simple_clean)
mv = mv[mv["text_clean"].str.len() > 0].reset_index(drop=True)
mv.head(3)
```

```python
!pip install textblob
!python -m textblob.download_corpora
from textblob import TextBlob
import seaborn as sns
import matplotlib.pyplot as plt

def textblob_polarity(text):
    try:
        return TextBlob(text).sentiment.polarity
    except Exception:
        return 0.0

mv["polarity"] = mv["text"].map(textblob_polarity)

fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(data=mv, x="polarity", bins=40, kde=False, ax=ax)
ax.set_title("movie reviews polarity")
ax.set_xlabel("polarity")
ax.set_ylabel("count")
plt.tight_layout()
plt.show()
```

#### Analysis: The movie reviews dataset includes short pieces of writing labeled as positive or negative, but the histogram of sentiment scores looks almost like a normal bell curve. This shape tells us that most reviews sit near the middle of the sentiment scale, meaning they contain a mix of positive and negative words.

#### The slight lean to the right shows that reviewers tend to use slightly more positive language overall. Because TextBlob calculates an average score for each review, it smooths out extreme opinions, which is why the results form one balanced curve rather than two separate clusters.

## Dataset 2: Reuters

```python
from nltk.corpus import reuters

# fetch if missing
for pkg in ["reuters", "punkt"]:
    try:
        nltk.data.find(f"corpora/{pkg}")
    except LookupError:
        nltk.download(pkg)

print("reuters files:", len(reuters.fileids()))
```

##### Create Dataframe and Clean

```python
import pandas as pd
from collections import Counter

primary_topics = []
for fid in reuters.fileids():
    cats = reuters.categories(fid)
    if not cats:
        continue
    primary_topics.append(cats[0])

top_counts = Counter(primary_topics).most_common(6)
top_topics = [t for t, _ in top_counts]

rows = []
per_label_cap = 300
per_label_seen = {t: 0 for t in top_topics}

for fid in reuters.fileids():
    cats = reuters.categories(fid)
    if not cats:
        continue
    label = cats[0]
    if label not in top_topics:
        continue
    if per_label_seen[label] >= per_label_cap:
        continue
    text = reuters.raw(fid)
    rows.append({"text": text, "label": label, "dataset": "reuters"})
    per_label_seen[label] += 1

rt = pd.DataFrame(rows)
len(rt), {k: v for k, v in per_label_seen.items()}
```

```python
import string
import re
from nltk.corpus import stopwords

nltk.download("stopwords")
stop_words = set(stopwords.words("english"))
extra_stops = {"said", "mr", "mrs", "also", "one", "two"}
stop_words.update(extra_stops)

def simple_clean(text):
    if not isinstance(text, str):
        return ""
    t = text.lower()
    t = t.translate(str.maketrans({c: " " for c in string.punctuation}))
    t = t.translate(str.maketrans({c: " " for c in string.digits}))
    t = re.sub(r"\s+", " ", t).strip()
    words = [w for w in t.split() if w not in stop_words]
    return " ".join(words)

rt["text_clean"] = rt["text"].map(simple_clean)
rt = rt[rt["text_clean"].str.len() > 0].reset_index(drop=True)
```

```python
from collections import Counter
import plotly.express as px

def top_k_terms(series, k=15):
    c = Counter()
    for doc in series:
        c.update(doc.split())
    if not c:
        return pd.DataFrame(columns=["term", "count"])
    terms, counts = zip(*c.most_common(k))
    return pd.DataFrame({"term": terms, "count": counts})

label_top = []
for lbl, sub in rt.groupby("label", sort=True):
    df_top = top_k_terms(sub["text_clean"], k=15)
    df_top["label"] = lbl
    label_top.append(df_top)

label_top = pd.concat(label_top, ignore_index=True)

px.bar(
    label_top,
    x="term",
    y="count",
    color="label",
    title="Top Words per Reuters Topic"
).update_layout(xaxis_tickangle=30).show()
```

#### Analysis: This chart highlights the main words for each Reuters news topic. Words like oil and barrel appear in “crude,” while currency and rates show up in “money-fx,” helping show which terms define each subject.

## Dataset 3: Inaugural Addresses

```python
import nltk
from nltk.corpus import inaugural

try:
    nltk.data.find("corpora/inaugural")
except LookupError:
    nltk.download("inaugural")

print("inaugural files:", len(inaugural.fileids()))
```

```python
import pandas as pd
import string
import re

rows = []
for fid in inaugural.fileids():
    text = inaugural.raw(fid)
    year = fid.split("-")[0].split()[0]
    rows.append({"text": text, "label": year, "dataset": "inaugural"})

ig = pd.DataFrame(rows)

def simple_clean(text):
    if not isinstance(text, str):
        return ""
    t = text.lower()
    t = t.translate(str.maketrans({c: " " for c in string.punctuation}))
    t = t.translate(str.maketrans({c: " " for c in string.digits}))
    t = re.sub(r"\s+", " ", t).strip()
    return t

ig["text_clean"] = ig["text"].map(simple_clean)
ig = ig[ig["text_clean"].str.len() > 0].reset_index(drop=True)
```

```python
!pip install wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

full_text = " ".join(ig["text_clean"].tolist())

wc = WordCloud(width=1200, height=600, background_color="white").generate(full_text)
plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("Inaugural Addresses Word Cloud")
plt.tight_layout()
plt.show()
```

#### Analysis: The word cloud highlights key themes repeated throughout U.S. presidential inaugural speeches. Larger words like people, nation, and power show the main ideas presidents emphasize when addressing the country.
