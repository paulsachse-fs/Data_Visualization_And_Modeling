## Data

### Global Electric Vehicle Trends
https://www.kaggle.com/datasets/sonawanelalitsunil/global-electric-vehicle-trends

```python
import pandas as pd
df = pd.read_csv('Electric_Vehicle_Population_Data.csv')
df.head()
```

## Data Summary

This dataset tracks the adoption and growth of electric vehicles across different regions and countries. It includes details such as the number of EVs registered, the types of vehicles (cars, buses, and two-wheelers), growth rates, and overall market share. Some records also include information about charging infrastructure, which gives additional insight into how EV accessibility is expanding globally. The data comes from public records and global energy and transport reports compiled to show long-term changes in EV adoption. This dataset is interesting because it helps visualize how clean energy policies, consumer demand, and technology improvements are shaping the future of sustainable transportation worldwide.

## Interesting Questions

1. Which manufacturers have the most EVs in this dataset?

2. How has the number of EVs changed by year?

3. What is the average electric range by manufacturer?

4. How do counts differ by electric vehicle type?

5. Which locations have the most EVs?

### 1. Which manufacturers have the most EVs in this dataset?

```python
top_makes = df['Make'].value_counts()
top_makes.head()
```

#### Answer: Tesla has the most EVs in the set, followed distantly by Chevrolet, Nissan, Ford, and Kia.

### 2. How has the number of EVs changed by year?

```python
year_counts = df['Model Year'].value_counts().sort_index()
year_counts = year_counts.loc[year_counts.index != 2026]
year_counts.head()
```

```python
year_counts.tail()
```

```python
year_counts.plot(kind='line', title='EV Counts by Model Year (Excluding 2026)')
```

#### Answer: The data shows consistent growth in EV registrations each year, with a strong upward trend through 2023. There is a slight drop-off after 2023, suggesting fewer entries recorded so far for the newer years. The model year 2026 is excluded because it is still in progress.

### 3. What is the average electric range by manufacturer?

```python
nonzero = df.loc[df['Electric Range'] > 0]
avg_range_by_make_nonzero = (
    nonzero.groupby('Make', dropna=False)['Electric Range']
           .mean()
           .sort_values(ascending=False)
)
avg_range_by_make_nonzero.head()
```

```python
avg_range_by_make_nonzero.plot(kind='bar', title='Average Electric Range by Make (Electric Range > 0)')
```

#### Answer: When I exclude zero values as likely unknown/missing, Tesla is at the top with an average electric range ≈ 241 miles, followed by Jaguar with 234 and Polestar with 233.

### 4. How do counts differ by electric vehicle type?

```python
type_counts = df['Electric Vehicle Type'].value_counts()
type_counts.head()
```

```python
type_counts.plot(kind='bar', title='Counts by Electric Vehicle Type')
```

#### Answer: Battery Electric Vehicles (BEVs) appear much more frequently than Plug-in Hybrid Electric Vehicles (PHEVs) in this dataset. This suggests that fully electric models now make up the majority of registered EVs, reflecting stronger market adoption of all-electric vehicles.

### 5. Which locations have the most EVs?

```python
top_cities = df['City'].value_counts().head(20)
top_cities
```

```python
top_cities.plot(kind='bar', title='Top 20 Cities by EV Count')
```

#### Answer: Showing the top 20 cities gives a clear picture of where EV adoption is highest. The leading cities are large metropolitan areas with strong charging infrastructure and state incentives. These locations tend to have higher population density and greater access to EV-related resources, which supports higher registration counts.
