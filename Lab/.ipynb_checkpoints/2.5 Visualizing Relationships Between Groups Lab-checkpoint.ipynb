## Data Import

```python
import pandas as pd
import matplotlib.pyplot as plt

loans = pd.read_csv('loan_data.csv')
loans.head()
```

***

## Education and Loan Status

```python
pd.crosstab(loans['person_education'], loans['loan_status'])
```

```python
pd.crosstab(loans['person_education'], loans['loan_status']).plot(kind='bar', stacked=True)
plt.title('Loan Status by Education')
plt.xlabel('Education Level')
plt.ylabel('Number of Applicants')
plt.legend(title='Loan Status')
plt.show()
```

```python
pd.crosstab(loans['person_education'], loans['loan_status'], normalize='index')*100
```

##### Education Findings: Approval rates are very similar across all education levels, with each category around 10 to 11 percent. The stacked bar chart shows that applicants with high school, associate, bachelor’s, master’s, or doctorate degrees all experience roughly the same approval and denial proportions.

Education level does not appear to have a meaningful impact on loan approval. The model likely relies more on financial variables rather than educational background when making decisions.

***

## Home Ownership and Loan Status

```python
pd.crosstab(loans['person_home_ownership'], loans['loan_status'])
```

```python
pd.crosstab(loans['person_home_ownership'], loans['loan_status']).plot(kind='bar', stacked=True)
plt.title('Loan Status by Home Ownership')
plt.xlabel('Home Ownership Type')
plt.ylabel('Number of Applicants')
plt.legend(title='Loan Status')
plt.show()
```

```python
pd.crosstab(loans['person_home_ownership'], loans['loan_status'], normalize='index')*100
```

##### Home Ownership Findings: Home ownership shows a clear difference in outcomes. Applicants who own their homes are approved about 24 percent of the time, while renters are approved about 8 percent of the time. Applicants with a mortgage are almost never approved, with an approval rate close to 1 percent.

Home ownership status is one of the strongest factors influencing loan approval. Applicants who fully own their homes are viewed as more financially stable and are therefore more likely to be approved.

## Loan Intent and Loan Status

```python
pd.crosstab(loans['loan_intent'], loans['loan_status'])
```

```python
pd.crosstab(loans['loan_intent'], loans['loan_status']).plot(kind='bar', stacked=True)
plt.title('Loan Status by Loan Intent')
plt.xlabel('Loan Purpose')
plt.ylabel('Number of Applicants')
plt.legend(title='Loan Status')
plt.show()
```

##### Loan Intent Findings: Loan purpose has a noticeable impact on approval rates. Applicants applying for education or home improvement loans were approved about 18 to 19 percent of the time, while those applying for medical, personal, or debt consolidation loans were approved only about 6 to 7 percent of the time.

Loan intent appears to matter. Some loan types, like education or home improvement, may be seen as more productive or less risky uses of credit, while others such as medical or personal loans might signal higher risk or financial hardship.

## Previous Loan Defaults and Loan Status

```python
pd.crosstab(loans['previous_loan_defaults_on_file'], loans['loan_status'])
```

```python
pd.crosstab(loans['previous_loan_defaults_on_file'], loans['loan_status']).plot(kind='bar', stacked=True)
plt.title('Loan Status by Prior Default on File')
plt.xlabel('Prior Default on File')
plt.ylabel('Number of Applicants')
plt.legend(title='Loan Status')
plt.show()
```

```python
pd.crosstab(loans['previous_loan_defaults_on_file'], loans['loan_status'], normalize='index')*100
```

##### Previous Loan Defaults and Loan Status Findings: The data shows a very strong relationship. Applicants with no previous defaults had an approval rate around 12 percent, while those with a default on file had a zero percent approval rate.

This is the most decisive factor in the dataset. The model appears to automatically reject anyone who has defaulted on a previous loan. From a credit risk perspective, this is understandable, although it also indicates that the model follows strict rules rather than flexible evaluation.

## Age and Loan Status

```python
approved_age = loans[loans['loan_status']==1]['person_age'].dropna()
denied_age = loans[loans['loan_status']==0]['person_age'].dropna()
```

```python
plt.hist(denied_age, bins=20, alpha=0.6, label='0 = Denied')
plt.hist(approved_age, bins=20, alpha=0.6, label='1 = Approved')
plt.title('Age Distribution by Loan Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
plt.boxplot([approved_age, denied_age], tick_labels=['Approved (1)','Denied (0)'])
plt.title('Age by Loan Status (Box Plot)')
plt.ylabel('Age')
plt.show()
```

##### Age and Loan Status Findings: The box plot shows that applicants who were approved for loans tend to be younger overall. The median age for approved applicants is lower than for denied applicants, and the age range is narrower. Denied applicants cover a wider spread, with many in higher age ranges.

Age appears to have some influence on loan approval. Younger applicants are approved more frequently, while older applicants are more often denied. The difference is visible, though not extreme, suggesting that age may be a minor factor in the model’s decision-making process.

***

## Income and Loan Status

```python
approved_inc = loans[loans['loan_status']==1]['person_income'].dropna()
denied_inc = loans[loans['loan_status']==0]['person_income'].dropna()
```

```python
plt.hist(denied_inc, bins=20, alpha=0.6, label='0 = Denied')
plt.hist(approved_inc, bins=20, alpha=0.6, label='1 = Approved')
plt.title('Income Distribution by Loan Status')
plt.xlabel('Annual Income')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
plt.boxplot([approved_inc, denied_inc], tick_labels=['Approved (1)','Denied (0)'])
plt.title('Income by Loan Status (Box Plot)')
plt.ylabel('Annual Income')
plt.show()
```

##### Income and Loan Status Findings: The income visualizations show that applicants with higher incomes are more likely to be approved. The box plot indicates that the median income for approved applicants is noticeably higher than for those who were denied.

Income has a clear relationship with loan approval. Higher income applicants are more likely to be approved, which aligns with standard lending practices since income reflects an applicant’s ability to repay.

The visualizations show that several financial and demographic factors influence loan approval decisions, while others appear to have little to no effect. The model most strongly considers whether an applicant has a prior default on file, as every applicant with a previous default was denied. Home ownership also plays a major role, with people who fully own their homes being approved far more often than renters or those with mortgages. Loan purpose and income show moderate relationships to approval, as applicants seeking education or home improvement loans and those with higher incomes are more likely to be approved. Age shows a small pattern, with younger applicants approved slightly more often, while education level and gender appear to have minimal impact. Overall, the model relies heavily on financial stability indicators and does not show major signs of gender bias, though the slight difference by age may warrant further review.
