## Setup

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

df = pd.read_csv('Employee_productivity.csv')
df.info()
df.head()
```

***
# Part 1: Determining a Strong Linear Relationship

```python
sns.regplot(x='YearsAtCompany', y='ProductivityScore', data=df)
plt.title("Productivity Score and Years at Company")
plt.show()

X = sm.add_constant(df['YearsAtCompany'])
y = df['ProductivityScore']
model = sm.OLS(y,X).fit()
model.summary()
```

```python
sns.regplot(x='HoursWorked', y='ProductivityScore', data=df)
plt.title("Productivity Score and Hours Worked")
plt.show()

X = sm.add_constant(df['HoursWorked'])
y = df['ProductivityScore']
model = sm.OLS(y,X).fit()
model.summary()
```

```python
sns.regplot(x='FocusScore', y='ProductivityScore', data=df)
plt.title("Productivity Score and Focus Score")
plt.show()

X = sm.add_constant(df['FocusScore'])
y = df['ProductivityScore']
model = sm.OLS(y,X).fit()
model.summary()
```

#### Part 1 Findings

The strongest linear relationship here belongs to that between the productivity score and focus score. The relationship between these items shows data points scattered evenly across both sides of the average line on the plot. Additionally, this relationship had the highest R-squared by a large margin at 0.678 and lowest P value at 0, while the other relationships had a R-squared < 0.1 and significant P values.

# Part 2: Additional Data

### Evaluating Department:
Based on the data below, we see that adding department into the data set does not change the R-squared value at all, indicating this factor does not greatly effect the model.

```python
df.loc[ df['Department'] == 'HR', 'dep_num'] = 0
df.loc[ df['Department'] == 'Marketing', 'dep_num'] = 1
df.loc[ df['Department'] == 'Sales', 'dep_num'] = 2
df.loc[ df['Department'] == 'Engineering', 'dep_num'] = 3

pd.crosstab(df['Department'], df['dep_num'])
```

```python
X = sm.add_constant(df[['FocusScore', "dep_num"]])
y = df['ProductivityScore']
model = sm.OLS(y,X).fit()
model.summary()
```

### Evaluating Team Meetings Per Week:
As we can see below, while the plot does suggest a slight downward trend, the data points are scattered too far from the trend line for it to be meaningful. Moreover, our r-squared value has not changed at all, signaling no effect from the added "team meetings per week" factor.

```python
pd.crosstab(df['Department'], df['TeamMeetingsPerWeek'])
```

```python
sns.regplot(x='TeamMeetingsPerWeek', y='ProductivityScore', data=df)
plt.title("Productivity Score and Team Meetings Per Week")
plt.show()
```

```python
X = sm.add_constant(df[['FocusScore', "TeamMeetingsPerWeek"]])
y = df['ProductivityScore']
model = sm.OLS(y,X).fit()
model.summary()
```

### Evaluating Desk Location:
Below, we can see that the r-squared value moves up slightly when adding in desk location as a factor, but the shift is not significant. I would not consider this a meaningful additional piece of data. Still, private offices appear to be slightly more productive than open plan offices, and remote work is the most productive according to this data set. 

```python
df.loc[ df['DeskLocation'] == 'Open Plan', 'desk_num'] = 0
df.loc[ df['DeskLocation'] == 'Private Office', 'desk_num'] = 1
df.loc[ df['DeskLocation'] == 'Remote', 'desk_num'] = 2

pd.crosstab(df['Department'], df['desk_num'])
```

```python
X = sm.add_constant(df[['FocusScore', "desk_num"]])
y = df['ProductivityScore']
model = sm.OLS(y,X).fit()
model.summary()
```

```python
sns.regplot(x='desk_num', y='ProductivityScore', data=df)
plt.title('Productivity Score and Desk Location')
plt.show()
```

# Part 3: Summary

According to this data set, by far the single most important variable in someone's productivity is their ability to focus. There was a clear linear trend when looking at the focus data and it has the highest R-squared value by itself at 0.678. From there, most other factors seem to make little to no difference, with a slight bump in productivity from people working in private offices or remote. My recommendation to this company would be to allow its employees to work in a private office or remote in order to improve their ability to focus.
